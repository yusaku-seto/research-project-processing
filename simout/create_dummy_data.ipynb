```python
# パラメータ設定
duration = 10.0  # シミュレーション時間（秒）
dt = 0.001  # 時間刻み
num_steps = int(duration / dt) + 1  # ステップ数

# 時系列データ
time = np.linspace(0, duration, num_steps)

# 加速度（ランダムな変動）
acceleration = np.random.uniform(-1, 1, num_steps)

# y方向の速度（加速度を積分）
velocity = np.cumsum(acceleration) * dt

# ヨー角（滑らかに変化させる）
yaw_rate = np.random.uniform(-0.02, 0.02, num_steps)  # 小さめの変動
yaw_angle = np.cumsum(yaw_rate) * dt

# x, y座標（速度とヨー角を考慮）
x_position = np.zeros(num_steps)
y_position = np.zeros(num_steps)
for i in range(1, num_steps):
    x_position[i] = x_position[i-1] + velocity[i] * np.cos(yaw_angle[i]) * dt
    y_position[i] = y_position[i-1] + velocity[i] * np.sin(yaw_angle[i]) * dt

# アクセル・ブレーキ量（加速時のみアクセル、減速時のみブレーキ）
accel_input = np.maximum(acceleration, 0)  # 加速時のみ値を持つ
brake_input = np.maximum(-acceleration, 0)  # 減速時のみ値を持つ

# データフレーム作成
df = pd.DataFrame({
    "simout1": time,
    "simout2": acceleration,
    "simout3": velocity,
    "simout4": x_position,
    "simout5": y_position,
    "simout6": yaw_angle,
    "simout7": accel_input,
    "simout8": brake_input
})

current_time = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")

# # CSVに保存
# df.to_csv("dummy_data.csv", index=False)
subject_id = "10"
if int(subject_id) % 2 == 1:
    experiment_type = "A"
else:
    experiment_type = "B"
file_name = f"output_{subject_id}_60_{experiment_type}_{current_time}.csv"
df.to_csv(file_name, index=False)

print("ダミーデータを dummy_data.csv に保存しました。")
df
```
